### List the most important variables (feature importance) that will affect the target label.

#### It appears some of the most important features that result in lower churn rate (in no particular order) include having: 1) a partner, 2) dependents, 3) longer tenures, 4) shorter contract types, and 5) non-paperless billing. Additional factors for lower churn rate include: 6) payment method (automatic bank transfers churn the least, followed by automatic credit card payments, electronic check payments, and mailed check payments which churn by far the most), and 7) which internet service is used (people with Fiber Optic are much more likely to churn than people with no internet service, while those with DSL are the least likely to churn).

## Step 5: State limitations / issues (if any) with the given dataset.

#### Many of the columns with binary answers (such as gender, senior citizen, phone service, multiple lines,  online security, online backup, device protection, tech support, streaming TV,  and streaming movies) were a bit difficult to work with. It may have been helpful if age was broken up into categories (i.e., <18, 18 - 25, 26 - 35, 36 - 45, 46 - 55, 56 - 65, 66 - 75, 76 - 85, >86). This way we could measure how gender might fluctuate in regards to the churn rate, depending on age. 
#### Total charges also created an issue at first, given that it was saved as an object data type rather than a float or integer.
#### More data could have also been helpful (i.e., state / country of residence, main spoken language, class, ethnicity, relgion, sexuality, etc.). This could be helpful in showing whether the demographic that did not churn skewed a certain way (for example, upper-class straight white Christian men between 30 - 45, who live in Texas). This is obviously a very specific example, but knowing such data could provide a lot of information (such as the possibility that Telco's services are better marketed towards people with more money, straight people, Christians, men, and/or people with southern values). This information could allow the company to either diversify their market or focus even more specifically on the demographic that they have already won over.

```python
from sklearn.model_selection import cross_val_score

models = []

def classification_Models(X2_train, X2_test, y2_train, y2_test ):

    models.append(("NB", GaussianNB()))
    models.append(("LR", LogisticRegression()))
    models.append(("RF", RandomForestClassifier()))
    models.append(("XGB", XGBClassifier()))
    
    modeloutcomes = []
    modelnames = []
    for name, model in models:
        v_results = cross_val_score(model, X_train, y_train, cv = 3, scoring = "recall", n_jobs = -1, verbose = 0)
        print(name, v_results.mean())
        modeloutcomes.append(v_results)
        modelnames.append(name)
        
    print(modeloutcomes)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xticklabels(modelnames)
    plt.boxplot(modeloutcomes)
        
classification_Models(X_train, X_test, y_train, y_test)
```
